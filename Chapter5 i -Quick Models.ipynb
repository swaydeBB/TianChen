## Ridge

```python
alphas = [0.00001, 0.0001, 0.0001, 0.001, 0.01, 0.1, 1.0, 10.0]
log_alphas = np.log10(alphas)

# 初始化Ridge模型
model = Ridge()

# 存储不同alpha对应的MSE值
mse_values = []

# 循环遍历不同的alpha值
for alpha in alphas:
    # 设置当前alpha值
    model.alpha = alpha
    
    # 使用GridSearchCV进行5折交叉验证
    grid_search = GridSearchCV(estimator=model, param_grid={}, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(x_train, y_train)
    
    # 获取负MSE值并取相反数得到MSE值
    mse = -grid_search.best_score_
    mse_values.append(mse)

# 绘制MSE曲线图（以log10的形式）
plt.figure(figsize=(10, 6))
plt.semilogx(alphas, mse_values, marker='o', linestyle='-')
plt.xlabel('Log10(Alpha)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Ridge Regression: MSE vs. Log10(Alpha)')
plt.grid(True)
plt.show()

# 找到最优的alpha值
best_alpha_index = np.argmin(mse_values)
best_alpha = alphas[best_alpha_index]
print(f"Best Alpha: {best_alpha}")
```

将超参数alpha的取值以log10的形式进行曲线绘制，可以看到0.1是最好的点。

```python
start_time = time.time()

ridge_model = Ridge(alpha=best_alpha)
ridge_model.fit(x_train, y_train)
ridge_y_pred = ridge_model.predict(x_test)

end_time = time.time()
ridge_total_time = end_time - start_time
print(f"Total Running Time: {ridge_total_time:.2f} seconds")
```

```python
# 创建Lasso模型
lasso_model = Lasso()

# 定义超参数网格
param_grid = {'alpha': [0.00001, 0.0001, 0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100, 1000]}

# 使用网格搜索来找到最佳的alpha超参数
grid_search = GridSearchCV(lasso_model, param_grid, cv=5)
grid_search.fit(x, y)
best_alpha = grid_search.best_params_['alpha']
print(f"Best Alpha: {best_alpha}")
```

```python


# 创建 Lasso 模型的 alpha 值候选列表
alphas = [0.00001, 0.0001, 0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100, 1000]

# 创建 LassoCV 模型，它可以自动执行交叉验证来选择最佳的 alpha
lasso_cv_model = LassoCV(alphas=alphas, cv=5)

# 训练模型
lasso_cv_model.fit(x_train, y_train)

# 将 alpha 值取对数
log_alphas = np.log10(lasso_cv_model.alphas_)

# 绘制肘图（对数刻度）
plt.figure(figsize=(10, 6))
plt.semilogx(10**log_alphas, lasso_cv_model.mse_path_, ':')
plt.plot(10**log_alphas, lasso_cv_model.mse_path_.mean(axis=-1), 'k', label='Average MSE Across Folds', linewidth=2)
plt.axvline(10**np.log10(lasso_cv_model.alpha_), linestyle='--', color='k', label='Best Alpha')
plt.xlabel('Alpha (log10 scale)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Lasso Regression: MSE vs. Alpha (log10 scale)')
plt.legend()
plt.grid(True)
plt.show()

# 打印最佳 alpha
print(f"Best Alpha: {lasso_cv_model.alpha_}")
```

结果保留四位小数，并且做成表格，并且加上贡献百分比，并且根据贡献比排序.贡献百分比这一栏挡在coef前面

Lasso回归模型在Sklearn中并没有像线性回归或岭回归那样的coef_属性来直接输出特征的系数。这是因为Lasso回归使用L1正则化，它会将一些特征的系数压缩至零，从而进行特征选择，因此不是所有特征都具有非零系数。

要获取Lasso回归模型中特征的贡献值，您可以采用以下步骤：

训练Lasso回归模型，并获得特征系数（系数可以是零或非零）。
计算特征的绝对值系数的和，以便计算特征的贡献度。
计算每个特征的贡献百分比，它等于该特征的绝对值系数除以所有特征绝对值系数之和的百分比。

```python
import matplotlib.pyplot as plt

# 训练 Lasso 模型
lasso_model = Lasso(alpha=best_alpha)
lasso_model.fit(x, y)

# 获取 Lasso 模型的系数
lasso_coef = lasso_model.coef_

# 计算特征贡献百分比
total_abs_coef = np.abs(lasso_coef).sum()

# 添加正负号并计算带有正负号的特征贡献百分比
lasso_coef_percentage_signed = (lasso_coef / total_abs_coef) * 100

# 将 lasso_coef 转换为一维数组
lasso_coef = lasso_coef.flatten()

# 创建一个 DataFrame 来存储特征、系数和带正负号的贡献百分比
lasso_coef_df = pd.DataFrame({'Feature': x.columns.tolist(), 'Coefficient': lasso_coef, 'Contribution (%)': lasso_coef_percentage_signed})

# 按特征带正负号的贡献百分比降序排序
lasso_coef_df = lasso_coef_df.sort_values(by='Contribution (%)', ascending=False)

# 打印 DataFrame
print(lasso_coef_df)


# 将特征重要性DataFrame按照贡献百分比升序排序
lasso_coef_df = lasso_coef_df.sort_values(by='Contribution (%)', ascending=True)

# 创建一个水平条形图
plt.figure(figsize=(8, 5))
bars = plt.barh(lasso_coef_df['Feature'], lasso_coef_df['Contribution (%)'], color='blue')

# 添加百分比标签
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}%', ha='left', va='center')

# 隐藏右边和顶部的边框
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# 设置图表标题
plt.title('Lasso Regression Feature Importance Ranking', fontsize=16)

# 紧凑布局
plt.tight_layout()

# 显示图表
plt.show()
```

```python
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score, max_error
from tqdm import tqdm  # 导入tqdm库用于显示进度条

# 初始化K近邻回归模型
KNN_regressor = KNeighborsRegressor(n_neighbors=10)  # 可根据需求设置参数

# 使用交叉验证进行预测
kf = KFold(n_splits=5, shuffle=True, random_state=42)
KNN_y_pred_cv = []  # 创建一个空列表来保存每个折的预测结果

for train_index, test_index in tqdm(kf.split(x), total=kf.get_n_splits(), desc="Cross-validation"):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    KNN_regressor.fit(x_train, y_train)
    KNN_y_pred_cv.extend(KNN_regressor.predict(x_test))  # 将每个折的预测结果添加到列表中

# 计算评估指标
knn_mse_cv = mean_squared_error(y, KNN_y_pred_cv)
knn_mae_cv = mean_absolute_error(y, KNN_y_pred_cv)
knn_r2_cv = r2_score(y, KNN_y_pred_cv)
knn_ev_cv = explained_variance_score(y, KNN_y_pred_cv)
knn_max_error_cv = max_error(y, KNN_y_pred_cv)

# 打印评估结果
print("Mean Squared Error (MSE): {:.10f}".format(knn_mse_cv))
print("Mean Absolute Error (MAE): {:.10f}".format(knn_mae_cv))
print("R-squared (R2): {:.10f}".format(knn_r2_cv))
print("Explained Variance (EV): {:.10f}".format(knn_ev_cv))
print("Max Error: {:.10f}".format(knn_max_error_cv))
```
