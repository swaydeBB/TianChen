```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from tqdm import tqdm
import itertools

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

# 创建GradientBoostingRegressor对象
GBR_regressor = GradientBoostingRegressor(random_state=42)

# 使用tqdm包装超参数组合的迭代过程
param_combinations = itertools.product(param_grid['n_estimators'], param_grid['learning_rate'], param_grid['max_depth'])
total_combinations = len(param_grid['n_estimators']) * len(param_grid['learning_rate']) * len(param_grid['max_depth'])

best_params = None
best_score = float("-inf")

for params in tqdm(param_combinations, total=total_combinations):
    n_estimators, learning_rate, max_depth = params
    GBR_regressor.set_params(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth)
    
    # 训练模型并评估性能，这里需要使用你的数据x和y
    GBR_regressor.fit(x, y)
    mse = -GBR_regressor.score(x, y)  # 根据需要更改性能评价指标
    
    if mse > best_score:
        best_score = mse
        best_params = params

print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

```python
import csv
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from tqdm import tqdm
import itertools
import numpy as np

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

# 创建GradientBoostingRegressor对象
GBR_regressor = GradientBoostingRegressor(random_state=42)

# 使用tqdm包装超参数组合的迭代过程
param_combinations = itertools.product(param_grid['n_estimators'], param_grid['learning_rate'], param_grid['max_depth'])
total_combinations = len(param_grid['n_estimators']) * len(param_grid['learning_rate']) * len(param_grid['max_depth'])

best_params = None
best_score = float("-inf")

# 创建一个列表来存储每个参数组合的指标
param_metrics = []

# 定义批处理大小
batch_size = 1000

# 获取样本数量
num_samples = x.shape[0]

for params in tqdm(param_combinations, total=total_combinations):
    n_estimators, learning_rate, max_depth = params
    GBR_regressor.set_params(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth)
    
    mse_total = 0.0
    mae_total = 0.0
    r_squared_total = 0.0
    me_total = 0.0
    
    for i in range(0, num_samples, batch_size):
        x_batch = x[i:i+batch_size]
        y_batch = y[i:i+batch_size]

        # 训练模型并预测
        GBR_regressor.fit(x_batch, y_batch)
        y_pred = GBR_regressor.predict(x_batch)

        mse_total += np.mean((y_pred - y_batch.values) ** 2)
        mae_total += np.mean(np.abs(y_pred - y_batch.values))
        r_squared_total += GBR_regressor.score(x_batch, y_batch.values)
        me_total += np.mean(y_pred - y_batch.values)
    
    mse = mse_total / (num_samples / batch_size)
    mae = mae_total / (num_samples / batch_size)
    r_squared = r_squared_total / (num_samples / batch_size)
    me = me_total / (num_samples / batch_size)
    
    # 存储指标到列表中
    param_metrics.append({
        'Parameters': str(params),
        'MSE': mse,
        'MAE': mae,
        'R^2': r_squared,
        'ME': me
    })
    
    if mse > best_score:
        best_score = mse
        best_params = params

print("Best Parameters:", best_params)
print("Best Score:", best_score)

# 输出每个参数组合的指标
for metrics in param_metrics:
    print("Parameters:", metrics['Parameters'])
    print("MSE:", metrics['MSE'])
    print("MAE:", metrics['MAE'])
    print("R^2:", metrics['R^2'])
    print("ME:", metrics['ME'])

# 将指标保存到CSV文件
with open('param_metrics.csv', mode='w', newline='') as csv_file:
    fieldnames = ['Parameters', 'MSE', 'MAE', 'R^2', 'ME']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    writer.writeheader()
    for metrics in param_metrics:
        writer.writerow(metrics)
```

## XGBR

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor

# 定义超参数搜索空间
param_grid = {
    'n_estimators': [50, 100, 200],  # 选择不同的树的数量
    'learning_rate': [0.01, 0.1, 0.2],  # 选择不同的学习率
    'max_depth': [3, 4, 5]  # 选择不同的树的最大深度
}

# 初始化XGBoost回归模型
XGB_regressor = XGBRegressor(random_state=42)

# 创建Grid Search对象，设置评分指标和交叉验证
grid_search = GridSearchCV(estimator=XGB_regressor, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)

# 执行Grid Search来搜索最佳超参数组合
grid_search.fit(x_train, y_train)

# 输出最佳超参数组合和对应的性能得分
best_params = grid_search.best_params_
best_score = -grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Negative MSE Score:", best_score)
```
