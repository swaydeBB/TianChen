## Read Datasets

```python
x = pd.read_csv('beijing_NM_x.csv', index_col=0,encoding="gbk")
y = pd.read_csv('beijing_NM_y.csv', index_col=0,encoding="gbk")
y = y['totalPrice']
x = pd.DataFrame(x)
y = pd.DataFrame(y)
x = x.drop(columns=['price','Cid'])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Models

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
import time

# 初始化K近邻回归模型，将K设置为7
KNN_regressor = KNeighborsRegressor(n_neighbors=7)

# 计时开始
start_time = time.time()

# 训练模型
KNN_regressor.fit(x_train, y_train)  # 请将x_train和y_train替换为你的训练数据

# 进行预测
KNN_y_pred = KNN_regressor.predict(x_test)  # 请将x_test替换为你的测试数据

# 计时结束
end_time = time.time()
KNN_total_time = end_time - start_time
print(f"Total Running Time: {KNN_total_time:.2f} seconds")
```

```python
KNN_regressor = KNeighborsRegressor(n_neighbors=7)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
KNN_y_pred_cv = cross_val_predict(KNN_regressor, x, y, cv=kf)

knn_mse_cv = mean_squared_error(y, KNN_y_pred_cv)
knn_mae_cv = mean_absolute_error(y, KNN_y_pred_cv)
knn_r2_cv = r2_score(y, KNN_y_pred_cv)
knn_ev_cv = explained_variance_score(y, KNN_y_pred_cv)
knn_max_error_cv = max_error(y, KNN_y_pred_cv)

print("{:.10f}".format(knn_mse_cv))
print("{:.10f}".format(knn_mae_cv))
print("{:.10f}".format(knn_r2_cv))
print("{:.10f}".format(knn_ev_cv))
print("{:.10f}".format(knn_max_error_cv))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, KNN_y_pred, alpha=0.5)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('KNN Regression: Scatter Plot of Actual vs. Predicted Prices')
plt.show()
# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual', linestyle='-')
plt.plot(KNN_y_pred, label='Predicted', linestyle='-')
plt.xlabel('Data Index')
plt.ylabel('Total Price')
plt.title('Distribution of Actual vs. Predicted Prices (KNN Regression)')
plt.legend()
plt.show()
```

在MLP模型中，通常没有直接计算特征重要性的方法，因为MLP是一种深度神经网络，其特征重要性不像一些传统的机器学习模型那样容易解释。通常，你可以通过查看权重值来推断特征的相对重要性。

以下是一个示例代码，用于查看MLP模型的权重值，并将其作为特征的重要性来排序。请注意，这不是一个精确的特征重要性度量，但可以用来理解哪些特征对模型的预测产生了更大的影响。

```python


# 获取MLP模型的权重
weights = MLP_regressor.coefs_

# 计算每个特征的权重和
feature_weights = [np.abs(layer).sum(axis=1) for layer in weights]

# 计算每个特征的总权重
total_weights = np.sum(feature_weights, axis=0)

# 创建一个DataFrame来存储特征和带百分比的特征权重
feature_weight_df = pd.DataFrame({'Feature Name': x_train.columns.tolist(), 'Weight': total_weights})

# 按特征权重降序排序
feature_weight_df = feature_weight_df.sort_values(by='Weight', ascending=False)

# 打印特征权重排序
print("Feature Weight Ranking:")
print(feature_weight_df.to_string(index=False))

# 创建水平条形图
plt.figure(figsize=(10, 6))
bars = plt.barh(feature_weight_df['Feature Name'], feature_weight_df['Weight'], color='blue')

# 添加权重标签
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}', ha='left', va='center')

# 隐藏右边和顶部的边框
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# 设置图表标题
plt.title('MLP Feature Weight Ranking', fontsize=16)

# 紧凑布局
plt.tight_layout()

# 显示图表
plt.show()
```
